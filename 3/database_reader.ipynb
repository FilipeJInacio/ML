```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from PIL import Image
import csv

i = 0
j = 0

with open('HAM10000_metadata', 'r') as csvfile:
    csvreader = csv.DictReader(csvfile)
    for row in csvreader:
        if row['dx'] == 'nv':
            r, g, b = Image.open("HAM10000_images_part_1/" + row['image_id'] + ".jpg").resize((28, 28)).split()

            r_matrix = list(r.getdata())
            g_matrix = list(g.getdata())
            b_matrix = list(b.getdata())

            r_matrix = [r_matrix[i * 28:(i + 1) * 28] for i in range(28)]
            g_matrix = [g_matrix[i * 28:(i + 1) * 28] for i in range(28)]
            b_matrix = [b_matrix[i * 28:(i + 1) * 28] for i in range(28)]
            
            r_matrix = np.expand_dims(r_matrix, axis=-1)
            g_matrix = np.expand_dims(g_matrix, axis=-1)
            b_matrix = np.expand_dims(b_matrix, axis=-1)
            
            rgb_matrices = np.concatenate([r_matrix, g_matrix, b_matrix], axis=-1)
            
            if i == 0:
                new_array_x = rgb_matrices.reshape(1, 28, 28, 3)
                i += 1
            else:    
                new_array_x = np.concatenate([new_array_x, rgb_matrices.reshape(1, 28, 28, 3)], axis=0)
                
            if j == 0:
                new_array_y = [0.]
                j += 1
            else:    
                new_array_y = np.append(new_array_y, 0.)
                
        if row['dx'] == 'mel':
            r, g, b = Image.open("HAM10000_images_part_1/" + row['image_id'] + ".jpg").resize((28, 28)).split()

            r_matrix = list(r.getdata())
            g_matrix = list(g.getdata())
            b_matrix = list(b.getdata())

            r_matrix = [r_matrix[i * 28:(i + 1) * 28] for i in range(28)]
            g_matrix = [g_matrix[i * 28:(i + 1) * 28] for i in range(28)]
            b_matrix = [b_matrix[i * 28:(i + 1) * 28] for i in range(28)]
            
            r_matrix = np.expand_dims(r_matrix, axis=-1)
            g_matrix = np.expand_dims(g_matrix, axis=-1)
            b_matrix = np.expand_dims(b_matrix, axis=-1)
            
            rgb_matrices = np.concatenate([r_matrix, g_matrix, b_matrix], axis=-1)
            
            if i == 0:
                new_array_x = rgb_matrices.reshape(1, 28, 28, 3)
                i += 1
            else:    
                new_array_x = np.concatenate([new_array_x, rgb_matrices.reshape(1, 28, 28, 3)], axis=0)
                
            if j == 0:
                new_array_y = [1.]
                j += 1
            else:    
                new_array_y = np.append(new_array_y, 1.)
            
            

```
