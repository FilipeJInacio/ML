```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.load("Xtrain_Classification1.npy")
y = np.load("ytrain_Classification1.npy")
x_project = np.load("Xtest_Classification1.npy")

x = x.reshape(-1, 28, 28, 3)
x_project = x_project.reshape(-1, 28, 28, 3)
x = np.concatenate((x, x_project), axis=0)

new_y = np.full(x.shape[0], -1)
```

Start

```python
data = np.load("dermamnist.npz")
data_x = np.concatenate((data["train_images"], data["val_images"], data["test_images"]), axis=0)
data_y = np.concatenate((data["train_labels"], data["val_labels"], data["test_labels"]), axis=0)

print(x.shape)
print(new_y.shape)
```

```python
for i,each_x in enumerate(x):
    for j,each_x2 in enumerate(data_x):
        if np.array_equal(each_x,each_x2):
            new_y[i] = data_y[j]
```
